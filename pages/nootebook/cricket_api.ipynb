## Fetch live match data

### Subtask:
Create a function to fetch live match data from the Cricbuzz API. Implement a caching mechanism with a time-to-live (TTL) to avoid making excessive API requests and respect the user's request limit.


**Reasoning**:
Define a function to fetch live match data from the Cricbuzz API with caching using the `@st.cache_data` decorator.



```python
@st.cache_data(ttl=60)
def fetch_live_matches(url, headers):
    """Fetches live match data from the Cricbuzz API with caching."""
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Raise an HTTPError for bad responses (4xx or 5xx)
        return response.json()
    except requests.exceptions.RequestException as e:
        st.error(f"Error fetching live match data: {e}")
        return None
```

```python
def display_live_matches(data):
    """Processes and displays live match data in the Streamlit app."""
    if data and 'typeMatches' in data:
        for match_type_entry in data['typeMatches']:
            if 'seriesMatches' in match_type_entry:
                for series_match_entry in match_type_entry['seriesMatches']:
                    if 'seriesAdWrapper' in series_match_entry and 'matches' in series_match_entry['seriesAdWrapper']:
                        for match in series_match_entry['seriesAdWrapper']['matches']:
                            if 'matchInfo' in match:
                                match_info = match['matchInfo']
                                st.subheader(f"{match_info.get('seriesName')} - {match_info.get('matchDesc')}")
                                st.write(f"**Format:** {match_info.get('matchFormat')}")
                                st.write(f"**State:** {match_info.get('state')}")
                                st.write(f"**Status:** {match_info.get('status')}")

                                team1_name = match_info.get('team1', {}).get('teamName')
                                team2_name = match_info.get('team2', {}).get('teamName')

                                if team1_name and team2_name:
                                    st.write(f"**Teams:** {team1_name} vs {team2_name}")
                                elif team1_name:
                                    st.write(f"**Team 1:** {team1_name}")
                                elif team2_name:
                                     st.write(f"**Team 2:** {team2_name}")


                                if 'matchScore' in match:
                                    st.write("**Score:**")
                                    if 'team1Score' in match['matchScore'] and 'inngs1' in match['matchScore']['team1Score']:
                                        score1 = match['matchScore']['team1Score']['inngs1']
                                        st.write(f"  {team1_name}: {score1.get('runs', 'N/A')}/{score1.get('wickets', 'N/A')} ({score1.get('overs', 'N/A')} Overs)")
                                    if 'team2Score' in match['matchScore'] and 'inngs1' in match['matchScore']['team2Score']:
                                        score2 = match['matchScore']['team2Score']['inngs1']
                                        st.write(f"  {team2_name}: {score2.get('runs', 'N/A')}/{score2.get('wickets', 'N/A')} ({score2.get('overs', 'N/A')} Overs)")
                                st.markdown("---")
    else:
        st.write("No live match data available.")
```

```python
# Filter players by role
batters = players[players['role'].str.contains('Batsman|WK-Batsman', na=False)]
bowlers = players[players['role'].str.contains('Bowler', na=False)]
all_rounders = players[players['role'].str.contains('Allrounder', na=False)]

# Calculate the number of players to sample from each category per team
# This is a simple approach; more complex logic might be needed for perfect distribution
total_players = 60
num_teams = players['teamId'].nunique()

# Determine how many players to aim for from each role per team
# Distribute the total sample size roughly equally among roles and teams
players_per_role_per_team = total_players // (3 * num_teams) # 3 roles: Batsman, Bowler, Allrounder
remaining_players = total_players % (3 * num_teams)

sampled_player_ids = []

# Sample from each team and role
for team_id in players['teamId'].unique():
    team_batters = batters[batters['teamId'] == team_id]
    team_bowlers = bowlers[bowlers['teamId'] == team_id]
    team_all_rounders = all_rounders[all_rounders['teamId'] == team_id]

    # Sample players from each category, ensuring we don't sample more than available
    sampled_batters = team_batters.sample(min(len(team_batters), players_per_role_per_team), random_state=42)
    sampled_bowlers = team_bowlers.sample(min(len(team_bowlers), players_per_role_per_team), random_state=42)
    sampled_all_rounders = team_all_rounders.sample(min(len(team_all_rounders), players_per_role_per_team), random_state=42)

    sampled_player_ids.extend(sampled_batters['id'].tolist())
    sampled_player_ids.extend(sampled_bowlers['id'].tolist())
    sampled_player_ids.extend(sampled_all_rounders['id'].tolist())

# If we still need more players to reach 60, sample randomly from the remaining players
if len(sampled_player_ids) < total_players:
    remaining_players_df = players[~players['id'].isin(sampled_player_ids)]
    additional_sample_size = total_players - len(sampled_player_ids)
    additional_players = remaining_players_df.sample(min(len(remaining_players_df), additional_sample_size), random_state=42)
    sampled_player_ids.extend(additional_players['id'].tolist())

# Ensure the final list has unique player IDs and is limited to 60
sampled_player_ids = list(set(sampled_player_ids))[:total_players]

print(f"Sampled {len(sampled_player_ids)} player IDs:")
print(sampled_player_ids)
```
